```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file_1 = "../1_Extract/Tornadoes_SPC_1950to2015.csv"
tornadoes_1 = pd.read_csv(csv_file_1)
tornadoes_1.head()
```

```python
df_columns_1 = ["om", "yr","date","time","tz","st", "mag", "inj", "fat","slat", "slon","elat","elon", "len","wid"]
tornadoes_transformed = tornadoes_1[df_columns_1].copy()
tornadoes_transformed.head()
```

```python
tornadoes_transformed.rename(columns={"om":"Tornado_ID",
                                      "yr":"Year",
                                     "date":"Date",
                                     "time": "Time",
                                     "tz":"Timezone",
                                     "st":"State",
                                     "mag":"Magnitude",
                                     "inj":"Injuries",
                                     "fat":"Fatalities",
                                     "slat":"Start_Lat",
                                     "slon":"Start_Long",
                                     "elat":"End_Lat",
                                     "elon":"End_Long",
                                     "len":"Length",
                                     "wid":"Width"})
```
